```python
số_nguyên_a = int(input("Nhập số nguyên đầu: "))
số_nguyên_b = int(input("Nhập số nguyên thứ hai: "))
sum_numbers = số_nguyên_a + số_nguyên_b
print(f"Tổng hai số vừa nhập: {sum_numbers}")
```

```python
r = float(input("Nhập bán kính r: "))
import math
S = math.pi * r ** 2
print(f"diện tích: {S}")
```

```python
số_nguyên = int(input("Nhập một số nguyên: "))
if số_nguyên % 2 == 0:
  print(f"số vừa nhâp: {số_nguyên} là số chẵn ")
else:
  print(f"số vừa nhập: {số_nguyên}, là số lẻ")
```

```python
n1 = int(input("Số nguyên thứ nhất: "))
n2 = int(input("Số nguyên thứ hai: "))
n3 = int(input("Số nguyên thứ ba: "))

numbermaxxinglmao = max(n1, n2, n3)
print(f"Số lớn nhất trong ba số vừa nhập là: {numbermaxxinglmao}")
```

```python
year = int(input("Nhập năm: "))
if year % 4 == 0 and year % 100 != 0:
  print("Năm vừa nhập là năm nhuận")
else:
  print("Năm vừa nhập không phải năm nhuận")
```

```python
n = int(input("Nhập một số nguyên dương bất kì: "))
print(n)
a1 = 0
for a2 in range(n + 1):
  a1 += a2

print(f"Tổng các số từ 0 đến {n} là: {a1}")
```
