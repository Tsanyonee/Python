```python
import math

a = float(input("nhập hệ số a:"))
b = float(input("nhập hệ số b:"))
c = float(input("nhập  hệ số c:"))

if a == 0:
    if b == 0:
        if c == 0:
            print("phương trình vô số nghiệm")
        else:
            print("phương trình vô nghiệm lmao")
    else:
        x = -c / b
        print(f"phương trình bậc nhất, có 1 nghiệm là: x = {x}")
else:
    delta = b**2 - 4*a*c
    print(f"Delta = {delta}")
    if delta > 0:
        x1 = ((-b + delta**0.5) / (2*a))
        x2 = ((-b - delta**0.5) / (2*a))
        print("phương trình có 2 nghiệm phân biệt")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print("phương trình có nghiệm kép x = {x}")
    else:
      delta < 0
      print("phương trình vô nghiệm trong tập số thực")
```
