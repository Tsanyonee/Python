```python
import math

n = int(input("nhập một số nguyên xác điịnh:"))
print(n)
a = 0
b = 0
while b < n:
  if b % 2 == 0:
    a += b
  b += 1
print(f"Tổng của các số chẵn nhỏ hơn {n} là: {a}")
```
